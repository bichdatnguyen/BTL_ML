```python
import numpy as np
import pandas as pd
from extinction import fitzpatrick99

def jurassic_park(flux, ebv, eff_wl):
    A_lambda = fitzpatrick99(eff_wl, ebv * 3.1)
    flux_ext = flux * 10**(A_lambda / 2.5)
    return flux_ext, A_lambda

filter_wl = {
    "u": np.array([3641]),
    "g": np.array([4704]),
    "r": np.array([6155]),
    "i": np.array([7504]),
    "z": np.array([8695]),
    "y": np.array([10056])
}

# --- Train ---
RealFluxTrain = []
A_lambda_list_Train = []

for idx, row in df_train.iterrows():
    eff_wl = filter_wl[row["Filter"]]
    flux_corr, A_lambda = jurassic_park(row["Flux"], row["EBV"], eff_wl)
    RealFluxTrain.append(float(flux_corr[0]))
    A_lambda_list_Train.append(float(A_lambda[0]))

df_train["RealFlux"] = RealFluxTrain
df_train["A_lambda"] = A_lambda_list_Train

# --- Test ---
RealFluxTest = []
A_lambda_list_Test = []

for idx, row in df_test.iterrows():
    eff_wl = filter_wl[row["Filter"]]
    flux_corr, A_lambda = jurassic_park(row["Flux"], row["EBV"], eff_wl)
    RealFluxTest.append(float(flux_corr[0]))
    A_lambda_list_Test.append(float(A_lambda[0]))

df_test["RealFlux"] = RealFluxTest
df_test["A_lambda"] = A_lambda_list_Test
```

## 1. Thống kê cơ bản liên quan tới RealFlux
Phần này trích xuất được những đặc trưng sau:
- {band}_mean = np.mean(Realflux)
- {band}_median = np.median(Realflux)
- {band}_std = np.std(Realflux)
- {band}_min = np.min(Realflux)
- {band}_max = np.max(Realflux)
- {band}_range = np.max(RealFlux) - np.min(RealFlux)
- {band}_skew = skew(RealFlux)
- {band}_kurtosis = kurtosis(RealFlux)
- {band}_slope
- {band}_time_to_peak


```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

bands = ['u','g','r','i','z','y']

def aggregate_flux(df, bands):
    features = []
    
    # Chuyển cột về float để tránh lỗi object
    df['RealFlux'] = df['RealFlux'].astype(float)
    df['Time (MJD)'] = df['Time (MJD)'].astype(float)
    
    # Group theo object_id
    for obj_id, group in df.groupby('object_id'):
        feat = {'object_id': obj_id}
        
        for band in bands:
            band_data = group[group['Filter'] == band]
            flux = band_data['RealFlux'].values
            time = band_data['Time (MJD)'].values

            if len(flux) == 0:
                # Nếu không có dữ liệu trong band này
                feat.update({f'{band}_mean': np.nan,
                             f'{band}_median': np.nan,
                             f'{band}_std': np.nan,
                             f'{band}_min': np.nan,
                             f'{band}_max': np.nan,
                             f'{band}_range': np.nan,
                             f'{band}_skew': np.nan,
                             f'{band}_kurtosis': np.nan,
                             f'{band}_slope': np.nan,
                             f'{band}_time_to_peak': np.nan})
                continue

            # Thống kê cơ bản
            feat[f'{band}_mean'] = np.mean(flux)
            feat[f'{band}_median'] = np.median(flux)
            feat[f'{band}_std'] = np.std(flux)
            feat[f'{band}_min'] = np.min(flux)
            feat[f'{band}_max'] = np.max(flux)
            feat[f'{band}_range'] = np.max(flux) - np.min(flux)
            feat[f'{band}_skew'] = skew(flux)
            feat[f'{band}_kurtosis'] = kurtosis(flux)
            
            # Độ dốc tổng thể flux theo thời gian
            if len(flux) > 1 and np.ptp(flux) > 0:  # ptp = max - min
                feat[f'{band}_slope'] = np.polyfit(time, flux, 1)[0]
                peak_idx = np.argmax(flux)
                feat[f'{band}_time_to_peak'] = time[peak_idx] - np.min(time)
            else:
                feat[f'{band}_slope'] = 0
                feat[f'{band}_time_to_peak'] = 0

        features.append(feat)
    
    return pd.DataFrame(features)

# Áp dụng cho train
train_features = aggregate_flux(df_train, bands)

# Áp dụng cho test
test_features = aggregate_flux(df_test, bands)

#print(train_features.head())
```

## 2. Màu và tỉ lệ màu

```python
def add_colors(df):
    # Chênh lệch flux (colors)
    df['g_r'] = df['g_mean'] - df['r_mean']
    df['r_i'] = df['r_mean'] - df['i_mean']
    df['i_z'] = df['i_mean'] - df['z_mean']
    df['u_g'] = df['u_mean'] - df['g_mean']
    df['y_z'] = df['y_mean'] - df['z_mean']
    
    # Flux ratios (tùy chọn, tránh chia cho 0)
    df['g_r_ratio'] = df['g_mean'] / (df['r_mean'] + 1e-8)
    df['r_i_ratio'] = df['r_mean'] / (df['i_mean'] + 1e-8)
    
    return df

# Áp dụng cho train/test
train_features = add_colors(train_features)
test_features = add_colors(test_features)
```

## 3. Thời lượng quan sát và số lần quan sát

```python
def add_temporal_features(df_original, df_feat):
    temporal_feats = []

    # Group theo object_id
    for obj_id, group in df_original.groupby('object_id'):
        feat = {
            'object_id': obj_id,
            'time_span': group['Time (MJD)'].max() - group['Time (MJD)'].min(),
            'num_obs': group.shape[0]  # tổng số điểm đo
        }
        temporal_feats.append(feat)
    
    temporal_df = pd.DataFrame(temporal_feats)
    df_feat = df_feat.merge(temporal_df, on='object_id', how='left')
    return df_feat

# Áp dụng cho train/test
train_features = add_temporal_features(df_train, train_features)
test_features = add_temporal_features(df_test, test_features)
```

## 4. RiseTime, Falltime tổng thể
(`rise_time`, `fall_time`, `asymmetry`, `peak_flux`)

```python
import pandas as pd
import numpy as np

def extract_time_features(df):
    features = []

    for obj_id, group in df.groupby('object_id'):
        feat = {'object_id': obj_id}

        # Peak flux và thời gian peak
        peak_idx = group['RealFlux'].idxmax()
        peak_time = group.loc[peak_idx, 'Time (MJD)']
        peak_flux = group.loc[peak_idx, 'RealFlux']
        feat['peak_flux'] = peak_flux
        feat['time_to_peak'] = peak_time - group['Time (MJD)'].min()

        # Rise time / fall time
        rise_time = peak_time - group['Time (MJD)'].min()
        fall_time = group['Time (MJD)'].max() - peak_time
        feat['rise_time'] = rise_time
        feat['fall_time'] = fall_time

        # Asymmetry
        feat['rise_fall_rate'] = rise_time / fall_time if fall_time > 0 else np.nan

        # Color at peak: flux ratios tại thời điểm peak
        flux_at_peak = group.loc[peak_idx]
        feat['g_r_at_peak'] = np.nan
        feat['r_i_at_peak'] = np.nan
        feat['i_z_at_peak'] = np.nan
        feat['u_g_at_peak'] = np.nan
        feat['y_z_at_peak'] = np.nan

        # Nếu các band có mặt tại thời điểm peak, tính ratio
        bands = group['Filter'].values
        fluxes = group['RealFlux'].values
        band_flux = dict(zip(bands, fluxes))
        for b1, b2, name in [('g','r','g_r_at_peak'),
                             ('r','i','r_i_at_peak'),
                             ('i','z','i_z_at_peak'),
                             ('u','g','u_g_at_peak'),
                             ('y','z','y_z_at_peak')]:
            if b1 in band_flux and b2 in band_flux and band_flux[b2] != 0:
                feat[name] = band_flux[b1] / band_flux[b2]
        
        features.append(feat)

    return pd.DataFrame(features)
```

```python
train_new_features = extract_time_features(df_train)
test_new_features  = extract_time_features(df_test)

train_features = train_features.merge(train_new_features, on='object_id', how='left')
test_features  = test_features.merge(test_new_features, on='object_id', how='left')
```

## 5. RiseTime, FallTime theo band

```python
import numpy as np
import pandas as pd

def build_risefall_features(df_train):
    """
    Tính rise/fall ratio per object_id & filter
    và trả về bảng phụ (object_id + các cột ratio_*).
    """

    results = []

    # ------------------------------------------------------------
    # Tính ratio cho từng object_id & Filter
    # ------------------------------------------------------------
    for (obj, filt), grp in df_train.groupby(["object_id", "Filter"]):

        grp = grp.sort_values("Time (MJD)")
        times = grp["Time (MJD)"].values
        flux  = grp["RealFlux"].values

        # Trường hợp toàn NaN
        if np.all(np.isnan(flux)):
            results.append((obj, filt, np.nan))
            continue

        # --- 1) Cực đại global ---
        peak_idx = np.nanargmax(flux)
        t_peak = times[peak_idx]

        # --- 2) Cực tiểu bên trái ---
        left_mask  = times < t_peak
        right_mask = times > t_peak

        if np.any(left_mask):
            t_left = times[left_mask].max()
        else:
            t_left = times.min()

        # --- 3) Cực tiểu bên phải ---
        if np.any(right_mask):
            t_right = times[right_mask].min()
        else:
            t_right = times.max()

        rise = t_peak - t_left
        fall = t_right - t_peak

        ratio = np.nan if fall == 0 else rise / fall

        results.append((obj, filt, ratio))

    # ------------------------------------------------------------
    # Convert → bảng rộng
    # ------------------------------------------------------------
    feat = (
        pd.DataFrame(results, columns=["object_id", "Filter", "rise_fall_ratio"])
          .pivot(index="object_id", columns="Filter", values="rise_fall_ratio")
    )

    # Đặt tên cột: ratio_u, ratio_g, ratio_r, ...
    feat = feat.add_prefix("ratio_")

    # Reset index để trả về object_id là cột
    feat = feat.reset_index()

    return feat
```

```python
rf_features = build_risefall_features(df_train)

train_features = train_features.merge(
    rf_features,
    on="object_id",
    how="left"
)

rf_features = build_risefall_features(df_test)

test_features = test_features.merge(
    rf_features,
    on="object_id",
    how="left"
)
```

## 6. Đặc trưng liên quan tới độ biến thiên thời gian và các giá trị RealFlux trên phân phối

```python
import numpy as np
import pandas as pd

def extract_basic_features(df):
    feats = {}

    # ----- TIME FEATURES -----
    times = df["Time (MJD)"].values
    feats["duration"] = times.max() - times.min()
    feats["time_mean"] = times.mean()
    feats["time_std"] = times.std()

    if len(times) > 1:
        time_diffs = np.diff(times)
        feats["time_gap_max"] = time_diffs.max()
        feats["time_gap_min"] = time_diffs.min()
    else:
        feats["time_gap_max"] = 0
        feats["time_gap_min"] = 0

    # ----- FLUX FEATURES -----
    flux = df["RealFlux"].values

    # percentiles
    feats["flux_p25"] = np.percentile(flux, 25)
    feats["flux_p75"] = np.percentile(flux, 75)
    feats["flux_p95"] = np.percentile(flux, 95)

    # flux steps (|Δflux|)
    if len(flux) > 1:
        flux_steps = np.abs(np.diff(flux))
        feats["flux_step_mean"] = flux_steps.mean()
        feats["flux_step_std"] = flux_steps.std()
    else:
        feats["flux_step_mean"] = 0
        feats["flux_step_std"] = 0

    # max slope = |Δflux / Δtime|
    if len(times) > 1:
        slopes = np.abs(np.diff(flux) / np.diff(times))
        feats["max_slope"] = slopes.max()
    else:
        feats["max_slope"] = 0

    return feats
```

```python
# ----- TRAIN NEW FEATURES -----
train_rows = []
for oid, g in df_train.groupby("object_id"):
    row = extract_basic_features(g)
    row["object_id"] = oid
    train_rows.append(row)

train_new = pd.DataFrame(train_rows)

# merge với train_features
train_features = train_features.merge(train_new, on="object_id", how="left")


# ----- TEST NEW FEATURES -----
test_rows = []
for oid, g in df_test.groupby("object_id"):
    row = extract_basic_features(g)
    row["object_id"] = oid
    test_rows.append(row)

test_new = pd.DataFrame(test_rows)

# merge với test_features
test_features = test_features.merge(test_new, on="object_id", how="left")
```

## 7. Trích xuất một số đặc trưng mạnh

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

def build_strong_features(df_train, time_tol=1.0):
    """
    Phiên bản rút gọn từ extract_strong_features:
    - Không merge vào df gốc
    - Trả về bảng phụ object-level: object_id + các strong features.
    """

    df = df_train.copy()
    bands = ["u","g","r","i","z","y"]

    # =====================================================
    # 1) Thống kê cơ bản per (object_id, Filter)
    # =====================================================
    stat = df.groupby(["object_id","Filter"])["RealFlux"].agg(
        mean="mean", std="std", min="min", max="max", median="median", count="count"
    ).reset_index()

    stat_pivot = stat.pivot(index="object_id", columns="Filter")
    stat_pivot.columns = [f"{metric}_{band}" for metric, band in stat_pivot.columns]

    # =====================================================
    # 2) amplitude & peak_norm (peak based on RealFlux)
    # =====================================================
    for b in bands:
        # amplitude
        if f"max_{b}" in stat_pivot.columns and f"min_{b}" in stat_pivot.columns:
            stat_pivot[f"amplitude_{b}"] = stat_pivot[f"max_{b}"] - stat_pivot[f"min_{b}"]
        else:
            stat_pivot[f"amplitude_{b}"] = np.nan
        
        # peak_norm = (peak - median) / std
        stat_pivot[f"peak_norm_{b}"] = np.where(
            (stat_pivot.get(f"std_{b}", np.nan)==0) | stat_pivot.get(f"std_{b}", np.nan).isna(),
            np.nan,
            (stat_pivot.get(f"max_{b}", np.nan) - stat_pivot.get(f"median_{b}", np.nan)) 
            / stat_pivot.get(f"std_{b}", np.nan)
        )

    # =====================================================
    # 3) rise_rate, fall_rate, rise_time, fall_time
    # =====================================================
    rates_rows = []
    grouped = df.groupby(["object_id","Filter"])

    for (obj, filt), grp in grouped:
        g = grp.sort_values("Time (MJD)")
        t = g["Time (MJD)"].values.astype(float)
        fvals = g["RealFlux"].values.astype(float)

        if np.all(np.isnan(fvals)):
            rates_rows.append((obj, filt, np.nan, np.nan, np.nan, np.nan))
            continue

        peak_idx = np.nanargmax(fvals)
        t_peak = t[peak_idx]
        f_peak = fvals[peak_idx]

        # left
        left_mask = t < t_peak
        if left_mask.any():
            t_left = t[left_mask].max()
            f_left = fvals[left_mask][-1]
        else:
            t_left = t.min()
            f_left = fvals[0]

        # right
        right_mask = t > t_peak
        if right_mask.any():
            t_right = t[right_mask].min()
            f_right = fvals[right_mask][0]
        else:
            t_right = t.max()
            f_right = fvals[-1]

        rise_time = t_peak - t_left
        fall_time = t_right - t_peak

        rise_rate = (f_peak - f_left) / rise_time if rise_time > 0 else np.nan
        fall_rate = (f_right - f_peak) / fall_time if fall_time > 0 else np.nan

        rates_rows.append((obj, filt, rise_rate, fall_rate, rise_time, fall_time))

    rates_df = pd.DataFrame(
        rates_rows, 
        columns=["object_id","Filter","rise_rate","fall_rate","rise_time","fall_time"]
    )
    rates_pivot = rates_df.pivot(index="object_id", columns="Filter")
    rates_pivot.columns = [f"{col}_{band}" for col, band in rates_pivot.columns]

    # =====================================================
    # 4) variance ratios (std ratios)
    # =====================================================
    var_feat = pd.DataFrame(index=stat_pivot.index)
    for b in bands:
        var_feat[f"var_std_{b}"] = stat_pivot.get(f"std_{b}", np.nan)

    var_feat["var_ratio_g_r"] = var_feat["var_std_g"] / var_feat["var_std_r"]
    var_feat["var_ratio_i_r"] = var_feat["var_std_i"] / var_feat["var_std_r"]
    var_feat["var_ratio_u_r"] = var_feat["var_std_u"] / var_feat["var_std_r"]
    var_feat["var_ratio_z_r"] = var_feat["var_std_z"] / var_feat["var_std_r"]
    var_feat["var_ratio_y_r"] = var_feat["var_std_y"] / var_feat["var_std_r"]

    # =====================================================
    # 5) color features: mean differences (RealFlux)
    # =====================================================
    color_df = pd.DataFrame(index=stat_pivot.index)
    for (b1, b2) in [("g","r"),("r","i"),("g","i")]:
        color_df[f"color_{b1}_{b2}"] = (
            stat_pivot.get(f"mean_{b1}", np.nan)
            - stat_pivot.get(f"mean_{b2}", np.nan)
        )

    # =====================================================
    # 6) color evolution slopes (using RealFlux)
    # =====================================================
    color_slope_records = []
    for obj, gobj in df.groupby("object_id"):
        gobj = gobj.sort_values("Time (MJD)")
        for (b1,b2) in [("g","r"),("r","i"),("g","i")]:
            s1 = gobj[gobj["Filter"]==b1][["Time (MJD)","RealFlux"]].rename(columns={"RealFlux":f"flux_{b1}"})
            s2 = gobj[gobj["Filter"]==b2][["Time (MJD)","RealFlux"]].rename(columns={"RealFlux":f"flux_{b2}"})
            if s1.empty or s2.empty:
                color_slope_records.append((obj, f"slope_color_{b1}_{b2}", np.nan))
                continue

            s1s2 = pd.merge_asof(
                s1.sort_values("Time (MJD)"), 
                s2.sort_values("Time (MJD)"),
                on="Time (MJD)", 
                direction="nearest", 
                tolerance=time_tol
            ).dropna()

            if len(s1s2) < 2:
                color_slope_records.append((obj, f"slope_color_{b1}_{b2}", np.nan))
                continue

            times = s1s2["Time (MJD)"].values.astype(float)
            colors = (s1s2[f"flux_{b1}"] - s1s2[f"flux_{b2}"]).values.astype(float)

            try:
                slope = LinearRegression().fit(times.reshape(-1,1), colors).coef_[0]
                color_slope_records.append((obj, f"slope_color_{b1}_{b2}", float(slope)))
            except:
                color_slope_records.append((obj, f"slope_color_{b1}_{b2}", np.nan))

    color_slope_df = pd.DataFrame(color_slope_records, columns=["object_id","feature","value"])
    color_slope_wide = (
        color_slope_df.pivot(index="object_id", columns="feature", values="value")
        if not color_slope_df.empty else 
        pd.DataFrame(index=stat_pivot.index)
    )

    # =====================================================
    # 7) asymmetry per band (RealFlux)
    # =====================================================
    asym_records = []
    for (obj, filt), grp in grouped:
        vals = grp["RealFlux"].dropna().values
        if len(vals)==0:
            asym = np.nan
        else:
            med = np.nanmedian(vals)
            above = vals[vals>med]
            below = vals[vals<med]
            asym = np.nan if len(above)==0 or len(below)==0 else (np.nanmean(above) - np.nanmean(below))
        asym_records.append((obj, filt, asym))

    asym_df = pd.DataFrame(asym_records, columns=["object_id","Filter","asym"])
    asym_wide = asym_df.pivot(index="object_id", columns="Filter", values="asym")
    asym_wide.columns = [f"asymmetry_{c}" for c in asym_wide.columns]

    # =====================================================
    # GỘP TẤT CẢ FEATURE
    # =====================================================
    features = (
        stat_pivot
        .join(rates_pivot, how="left")
        .join(var_feat, how="left")
        .join(color_df, how="left")
        .join(color_slope_wide, how="left")
        .join(asym_wide, how="left")
        .reset_index()
    )

    return features
```

```python
strong_feats_train = build_strong_features(df_train)

train_features = train_features.merge(
    strong_feats_train,
    on="object_id",
    how="left"
)

strong_feats_test = build_strong_features(df_test)

test_features = test_features.merge(
    strong_feats_test,
    on="object_id",
    how="left"
)
```

## 8. Trích xuất các đặc trưng sai số

```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

# --- Tạo các feature nâng cao cho từng object_id ---
def extract_flux_features(df):
    features = []

    grouped = df.groupby('object_id')
    
    for obj, group in grouped:
        feat = {'object_id': obj}
        
        # Flux_err cơ bản
        flux_err = group['Flux_err'].values
        feat['FluxErr_mean']   = np.nanmean(flux_err)
        feat['FluxErr_std']    = np.nanstd(flux_err)
        feat['FluxErr_min']    = np.nanmin(flux_err)
        feat['FluxErr_max']    = np.nanmax(flux_err)
        feat['FluxErr_median'] = np.nanmedian(flux_err)
        feat['FluxErr_skew']   = skew(flux_err, nan_policy='omit')
        feat['FluxErr_kurt']   = kurtosis(flux_err, nan_policy='omit')
        
        # WeightedFlux = RealFlux / Flux_err
        weighted_flux = group['WeightedFlux'].values
        feat['WeightedFlux_mean']   = np.nanmean(weighted_flux)
        feat['WeightedFlux_std']    = np.nanstd(weighted_flux)
        feat['WeightedFlux_min']    = np.nanmin(weighted_flux)
        feat['WeightedFlux_max']    = np.nanmax(weighted_flux)
        feat['WeightedFlux_median'] = np.nanmedian(weighted_flux)
        feat['WeightedFlux_skew']   = skew(weighted_flux, nan_policy='omit')
        feat['WeightedFlux_kurt']   = kurtosis(weighted_flux, nan_policy='omit')
        
        # Flux_err / RealFlux ratio
        flux_err_ratio = group['Flux_err_ratio'].values
        feat['FluxErrRatio_mean']   = np.nanmean(flux_err_ratio)
        feat['FluxErrRatio_std']    = np.nanstd(flux_err_ratio)
        feat['FluxErrRatio_min']    = np.nanmin(flux_err_ratio)
        feat['FluxErrRatio_max']    = np.nanmax(flux_err_ratio)
        feat['FluxErrRatio_median'] = np.nanmedian(flux_err_ratio)
        feat['FluxErrRatio_skew']   = skew(flux_err_ratio, nan_policy='omit')
        feat['FluxErrRatio_kurt']   = kurtosis(flux_err_ratio, nan_policy='omit')

        # Flux_err cơ bản
        flux_weight = group['Flux_weight'].values
        feat['Flux_weight_mean']   = np.nanmean(flux_weight)
        feat['Flux_weight_std']    = np.nanstd(flux_weight)
        feat['Flux_weight_min']    = np.nanmin(flux_weight)
        feat['Flux_weight_max']    = np.nanmax(flux_weight)
        feat['Flux_weight_median'] = np.nanmedian(flux_weight)
        feat['Flux_weight_skew']   = skew(flux_weight, nan_policy='omit')
        feat['Flux_weight_kurt']   = kurtosis(flux_weight, nan_policy='omit')
        
        # WeightedFlux = RealFlux / Flux_err
        log_Flux_err = group['Log_Flux_err'].values
        feat['Log_Flux_err_mean']   = np.nanmean(log_Flux_err)
        feat['Log_Flux_err_std']    = np.nanstd(log_Flux_err)
        feat['Log_Flux_err_min']    = np.nanmin(log_Flux_err)
        feat['Log_Flux_err_max']    = np.nanmax(log_Flux_err)
        feat['Log_Flux_err_median'] = np.nanmedian(log_Flux_err)
        feat['Log_Flux_err_skew']   = skew(log_Flux_err, nan_policy='omit')
        feat['Log_Flux_err_kurt']   = kurtosis(log_Flux_err, nan_policy='omit')
        
        # Flux_err / RealFlux ratio
        log_WeightedFlux = group['Log_WeightedFlux'].values
        feat['Log_WeightedFlux_mean']   = np.nanmean(log_WeightedFlux)
        feat['Log_WeightedFlux_std']    = np.nanstd(log_WeightedFlux)
        feat['Log_WeightedFlux_min']    = np.nanmin(log_WeightedFlux)
        feat['Log_WeightedFlux_max']    = np.nanmax(log_WeightedFlux)
        feat['Log_WeightedFlux_median'] = np.nanmedian(log_WeightedFlux)
        feat['Log_WeightedFlux_skew']   = skew(log_WeightedFlux, nan_policy='omit')
        feat['Log_WeightedFlux_kurt']   = kurtosis(log_WeightedFlux, nan_policy='omit')

        log_Flux_err_ratio = group['Log_Flux_err_ratio'].values
        feat['Log_Flux_err_ratio_mean']   = np.nanmean(log_Flux_err_ratio)
        feat['Log_Flux_err_ratio_std']    = np.nanstd(log_Flux_err_ratio)
        feat['Log_Flux_err_ratio_min']    = np.nanmin(log_Flux_err_ratio)
        feat['Log_Flux_err_ratio_max']    = np.nanmax(log_Flux_err_ratio)
        feat['Log_Flux_err_ratio_median'] = np.nanmedian(log_Flux_err_ratio)
        feat['Log_Flux_err_ratio_skew']   = skew(log_Flux_err_ratio, nan_policy='omit')
        feat['Log_Flux_err_ratio_kurt']   = kurtosis(log_Flux_err_ratio, nan_policy='omit')
        features.append(feat)
    
    return pd.DataFrame(features)

# --- Áp dụng ---
df_flux_features = extract_flux_features(df_train)

# Kiểm tra kết quả
df_flux_features.head()
```

## 9. Giữ lại các đặc trưng gốc

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np

def add_additional_features(df_original, df_feat, include_target=False):
    # Chuyển các cột số về float
    for col in ['Z', 'Z_err', 'EBV']:
        if col in df_original.columns:
            df_original[col] = pd.to_numeric(df_original[col], errors='coerce')
    
    # Loại bỏ các cột đã tồn tại trong df_feat để tránh MergeError
    for col in ['Z', 'Z_err', 'EBV', 'SpecType', 'target']:
        if col in df_feat.columns:
            df_feat = df_feat.drop(columns=[col])
    
    # Lấy các cột bổ sung theo object_id
    additional = []
    for obj_id, group in df_original.groupby('object_id'):
        feat = {
            'object_id': obj_id,
            'Z': group['Z'].iloc[0] if 'Z' in group.columns else np.nan,
            'Z_err': group['Z_err'].iloc[0] if 'Z_err' in group.columns else np.nan,
            'EBV': group['EBV'].iloc[0] if 'EBV' in group.columns else np.nan,
            'SpecType': str(group['SpecType'].iloc[0]) if 'SpecType' in group.columns else 'Unknown'
        }
        if include_target:
            feat['target'] = group['target'].iloc[0] if 'target' in group.columns else np.nan
        additional.append(feat)
    
    additional_df = pd.DataFrame(additional)
    
    # Merge vào feature table
    df_feat = df_feat.merge(additional_df, on='object_id', how='left')
    
    return df_feat

# Áp dụng
train_features = add_additional_features(df_train, train_features, include_target=True)
test_features = add_additional_features(df_test, test_features, include_target=False)
```

## LGBM

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, f1_score

#i=1
# 1. Chọn feature và target
X = pd.concat([train_features.drop(columns=cols_to_drop + ['target']), df_to_add_in_train], axis = 1)
y = train_features['target']

X_test = pd.concat([test_features.drop(columns=cols_to_drop, errors='ignore'), df_to_add_in_test], axis = 1)

# 2. KFold setup
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

# 3. LightGBM params
params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'learning_rate': 0.02,
    'num_leaves': 24, # 24 làm mốc
    'min_data_in_leaf': 50, # 100 làm mốc
    'feature_fraction': 0.8,
    'bagging_fraction': 0.6, # 0.6 làm mốc
    'bagging_freq': 5,
    'lambda_l1': 1.0,
    'lambda_l2': 2.0,
    'class_weight': 'balanced',
    'min_gain_to_split': 0.05, #0.05 làm mốc
    'max_depth': -1,
    'verbosity': -1,
    'seed': 0
}

# 4. Lưu kết quả
oof_pred = np.zeros(len(X))        
test_pred = np.zeros(len(X_test))
auc_scores = []
f1_scores_list = []
best_thresholds = []     # NEW: lưu threshold tốt nhất cho từng fold

# 5. Vòng lặp K-FOLD training
for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
    print(f"\n===== Fold {fold + 1} =====")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    dtrain = lgb.Dataset(X_train, label=y_train)
    dval = lgb.Dataset(X_val, label=y_val)

    callbacks = [
        lgb.early_stopping(stopping_rounds=500),
        lgb.log_evaluation(period=500)
    ]

    model = lgb.train(
        params,
        dtrain,
        num_boost_round=5000,
        valid_sets=[dval],
        valid_names=[f"val_fold_{fold+1}"],
        callbacks=callbacks
    )

    # Dự đoán
    val_pred = model.predict(X_val)
    oof_pred[val_idx] = val_pred
    test_pred += model.predict(X_test) / skf.n_splits

    # ===== NEW: Tìm threshold tối ưu cho fold này =====
    thresholds = np.linspace(0.01, 0.40, 200)
    best_f1 = 0
    best_t = 0

    for t in thresholds:
        f1 = f1_score(y_val, (val_pred >= t).astype(int))
        if f1 > best_f1:
            best_f1 = f1
            best_t = t

    best_thresholds.append(best_t)

    # Tính AUC & F1 theo threshold tối ưu
    auc_fold = roc_auc_score(y_val, val_pred)
    auc_scores.append(auc_fold)
    f1_scores_list.append(best_f1)

    print(f"Fold {fold+1} AUC       = {auc_fold:.4f}")
    print(f"Fold {fold+1} best_thr = {best_t:.4f}")
    print(f"Fold {fold+1} F1       = {best_f1:.4f}")

# 6. Kết quả trung bình
print("\n===== CV RESULT =====")
print(f"Mean AUC: {np.mean(auc_scores):.4f}  |  Std: {np.std(auc_scores):.4f}")
print(f"Mean F1 : {np.mean(f1_scores_list):.4f}  |  Std: {np.std(f1_scores_list):.4f}")
print(f"Per-fold thresholds:", best_thresholds)

# 7. OOF score
overall_auc = roc_auc_score(y, oof_pred)
print(f"\nOOF AUC (sanity check): {overall_auc:.4f}")

# 8. ===== NEW: tìm global threshold từ toàn bộ OOF =====
thresholds = np.linspace(0.01, 0.40, 400)
best_f1_global = 0
best_t_global = 0

for t in thresholds:
    f1 = f1_score(y, (oof_pred >= t).astype(int))
    if f1 > best_f1_global:
        best_f1_global = f1
        best_t_global = t

print(f"\nGLOBAL BEST THRESHOLD = {best_t_global:.4f}")
print(f"GLOBAL BEST F1        = {best_f1_global:.4f}")
#print(f"Phase                 = {i}")
#==============================================================================
#i+=1
#==============================================================================
# 9. Kết quả test cuối cùng (probability)
LGBM_y_pred_test = test_pred
```

```python
import numpy as np
import xgboost as xgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, f1_score

# 1. Feature/target
X = pd.concat([train_features.drop(columns=cols_to_drop + ['target']), df_to_add_in_train], axis = 1)
y = train_features['target']
X_test = pd.concat([test_features.drop(columns=cols_to_drop, errors='ignore'), df_to_add_in_test], axis = 1)

# 2. Stratified K-Fold
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=0)

# 3. Khởi tạo OOF và test prediction
oof_preds = np.zeros(len(X))
test_preds = np.zeros(len(X_test))
fold_best_thresholds = []
fold_f1_scores = []
fold_auc_scores = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y), 1):
    print(f"\n===== Fold {fold} =====")
    
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # FIX INF → NaN
    for df in [X_train, X_val, X_test]:
        df.replace([np.inf, -np.inf], np.nan, inplace=True)
    
    # DMatrix
    dtrain = xgb.DMatrix(X_train, label=y_train)
    dval   = xgb.DMatrix(X_val, label=y_val)
    dtest  = xgb.DMatrix(X_test)
    
    # Params
    params = {
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'eta': 0.02,
        'max_depth': 0,
        'min_child_weight': 3,
        'subsample': 0.78,
        'colsample_bytree': 0.49,
        'lambda': 0.5,
        'alpha': 0.25,
        'seed': 0,
    }
    
    # Train
    watchlist = [(dtrain, 'train'), (dval, f'val_fold_{fold}')]
    model_xgb = xgb.train(
        params,
        dtrain,
        num_boost_round=5000,
        evals=watchlist,
        early_stopping_rounds=500,
        verbose_eval=500
    )
    
    # Predict
    y_pred_val = model_xgb.predict(dval)
    y_pred_test = model_xgb.predict(dtest)
    
    # OOF & test
    oof_preds[val_idx] = y_pred_val
    test_preds += y_pred_test / n_splits
    
    # AUC
    auc_val = roc_auc_score(y_val, y_pred_val)
    fold_auc_scores.append(auc_val)
    
    # Find best threshold for F1
    thresholds = np.linspace(0.01, 0.99, 99)
    f1_scores = [f1_score(y_val, (y_pred_val >= t).astype(int)) for t in thresholds]
    best_idx = np.argmax(f1_scores)
    best_thr = thresholds[best_idx]
    best_f1 = f1_scores[best_idx]
    
    fold_best_thresholds.append(best_thr)
    fold_f1_scores.append(best_f1)
    
    print(f"Fold {fold} AUC       = {auc_val:.4f}")
    print(f"Fold {fold} best_thr = {best_thr:.4f}")
    print(f"Fold {fold} F1       = {best_f1:.4f}")

# CV summary
print("\n===== CV RESULT =====")
print(f"Mean AUC: {np.mean(fold_auc_scores):.4f}  |  Std: {np.std(fold_auc_scores):.4f}")
print(f"Mean F1 : {np.mean(fold_f1_scores):.4f}  |  Std: {np.std(fold_f1_scores):.4f}")
print(f"Per-fold thresholds: {fold_best_thresholds}")

#================================ Test ================================
#oof_preds = oof_preds*bias_sigma_exp
#======================================================================
# OOF AUC
oof_auc = roc_auc_score(y, oof_preds)
print(f"\nOOF AUC (sanity check): {oof_auc:.4f}")

# Global best threshold
global_best_idx = np.argmax([f1_score(y, (oof_preds >= t).astype(int)) for t in fold_best_thresholds])
global_best_thr = fold_best_thresholds[global_best_idx]
global_best_f1  = f1_score(y, (oof_preds >= global_best_thr).astype(int))
print(f"\nGLOBAL BEST THRESHOLD = {global_best_thr:.4f}")
print(f"GLOBAL BEST F1        = {global_best_f1:.4f}")

thresholds = np.linspace(0.01, 0.99, 99)
global_self_f1 = max([f1_score(y, (oof_preds >= t).astype(int)) for t in thresholds])

print(f"GLOBAL SELF F1        = {global_self_f1:.4f}")

XGB_y_pred_test = test_preds
```

```python
#================================== Lặp XGB ====================================
import numpy as np
import xgboost as xgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score, roc_auc_score

# --- Feature/target ---
X = pd.concat([train_features.drop(columns=cols_to_drop + ['target']), df_to_add_in_train], axis = 1)
y = train_features['target']
X_test = pd.concat([test_features.drop(columns=cols_to_drop, errors='ignore'), df_to_add_in_test], axis = 1)

# --- Các random_state cần thử nghiệm ---
seed_list = range(0, 51)
best_seed = None
best_f1_mean = -1
best_global_f1 = -1
best_global_thr = None

for seed in seed_list:
    print(f"\n===== Testing seed = {seed} =====")
    
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
    fold_f1_scores = []
    
    oof_preds = np.zeros(len(X))
    fold_best_thresholds = []
    
    for fold, (train_idx, val_idx) in enumerate(skf.split(X, y), 1):
        X_train, X_val = X.iloc[train_idx].copy(), X.iloc[val_idx].copy()
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        
        # FIX INF → NaN
        X_train.replace([np.inf, -np.inf], np.nan, inplace=True)
        X_val.replace([np.inf, -np.inf], np.nan, inplace=True)
        
        # DMatrix
        dtrain = xgb.DMatrix(X_train, label=y_train)
        dval   = xgb.DMatrix(X_val, label=y_val)
        
        # Params
        params = {
            'objective': 'binary:logistic',
            'eval_metric': 'logloss',
            'eta': 0.02,
            'max_depth': 0,
            'min_child_weight': 3,
            'subsample': 0.78,
            'colsample_bytree': 0.49,
            'lambda': 0.5,
            'alpha': 0.25,
            'seed': seed,
        }
        
        watchlist = [(dtrain, 'train'), (dval, f'val_fold_{fold}')]
        model_xgb = xgb.train(
            params,
            dtrain,
            num_boost_round=5000,
            evals=watchlist,
            early_stopping_rounds=500,
            verbose_eval=False
        )
        
        y_pred_val = model_xgb.predict(dval)
        
        # Tìm best threshold cho fold
        thresholds = np.linspace(0.01, 0.99, 99)
        f1_scores = [f1_score(y_val, (y_pred_val >= t).astype(int)) for t in thresholds]
        best_idx = np.argmax(f1_scores)
        best_thr = thresholds[best_idx]
        best_f1_fold = f1_scores[best_idx]
        
        fold_f1_scores.append(best_f1_fold)
        fold_best_thresholds.append(best_thr)
        
        # OOF
        oof_preds[val_idx] = y_pred_val
    
    mean_f1 = np.mean(fold_f1_scores)
    std_f1 = np.std(fold_f1_scores)
    
    # Global best F1 trên toàn bộ OOF
    global_f1_scores = [f1_score(y, (oof_preds >= t).astype(int)) for t in fold_best_thresholds]
    global_best_idx = np.argmax(global_f1_scores)
    global_best_thr = fold_best_thresholds[global_best_idx]
    global_best_f1 = global_f1_scores[global_best_idx]
    
    print(f"Seed {seed} -> Mean F1: {mean_f1:.4f}, Std F1: {std_f1:.4f}, Global Best F1: {global_best_f1:.4f} at threshold {global_best_thr:.4f}")
    
    if mean_f1 > best_f1_mean:
        best_f1_mean = mean_f1
        best_seed = seed
        best_global_f1_final = global_best_f1
        best_global_thr_final = global_best_thr
        best_f1_std_final = std_f1

print(f"\n===== Best seed = {best_seed} =====")
print(f"Mean F1 = {best_f1_mean:.4f}, Std F1 = {best_f1_std_final:.4f}")
print(f"Global Best F1 = {best_global_f1_final:.4f} at threshold {best_global_thr_final:.4f}")
```

```python
#====================== Tìm best random_states cho LGBM (tối ưu F1) ======================
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score

# --- Feature/target ---
X = pd.concat(
    [train_features.drop(columns=cols_to_drop + ['target']),
     df_to_add_in_train],
    axis=1
)
y = train_features['target']

X_test = pd.concat(
    [test_features.drop(columns=cols_to_drop, errors='ignore'),
     df_to_add_in_test],
    axis=1
)

# --- Các random_state cần thử ---
seed_list = range(0, 51)

best_seed = None
best_f1_mean = -1
best_f1_std = None
best_global_f1_final = -1
best_global_thr_final = None

for seed in seed_list:
    print(f"\n===== Testing seed = {seed} =====")

    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

    fold_f1_scores = []
    fold_best_thresholds = []
    oof_pred = np.zeros(len(X))

    # --------- 5-Fold CV ----------
    for fold, (train_idx, val_idx) in enumerate(skf.split(X, y), 1):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

        dtrain = lgb.Dataset(X_train, label=y_train)
        dval   = lgb.Dataset(X_val, label=y_val)

        params = {
            'objective': 'binary',
            'metric': 'binary_logloss',
            'learning_rate': 0.02,
            'num_leaves': 24,
            'min_data_in_leaf': 100,
            'feature_fraction': 0.4,
            'bagging_fraction': 0.6,
            'bagging_freq': 5,
            'lambda_l1': 1.0,
            'lambda_l2': 2.0,
            'min_gain_to_split': 0.05,
            'max_depth': -1,
            'class_weight': 'balanced',
            'verbosity': -1,
            'seed': seed
        }

        callbacks = [
            lgb.early_stopping(stopping_rounds=500),
            lgb.log_evaluation(period=0)
        ]

        model = lgb.train(
            params,
            dtrain,
            num_boost_round=5000,
            valid_sets=[dval],
            valid_names=[f"val_fold_{fold}"],
            callbacks=callbacks
        )

        # ---- Predict ----
        val_pred = model.predict(X_val)
        oof_pred[val_idx] = val_pred

        # ---- Tìm threshold tối ưu F1 cho fold ----
        thresholds = np.linspace(0.01, 0.40, 200)
        f1_scores = [
            f1_score(y_val, (val_pred >= t).astype(int))
            for t in thresholds
        ]

        best_idx = np.argmax(f1_scores)
        best_thr = thresholds[best_idx]
        best_f1  = f1_scores[best_idx]

        fold_f1_scores.append(best_f1)
        fold_best_thresholds.append(best_thr)

    # ---- F1 mean & std theo fold ----
    mean_f1 = np.mean(fold_f1_scores)
    std_f1  = np.std(fold_f1_scores)

    # ---- Global best F1 trên toàn bộ OOF ----
    global_f1_scores = [
        f1_score(y, (oof_pred >= t).astype(int))
        for t in fold_best_thresholds
    ]

    global_best_idx = np.argmax(global_f1_scores)
    global_best_thr = fold_best_thresholds[global_best_idx]
    global_best_f1  = global_f1_scores[global_best_idx]

    print(
        f"Seed {seed} -> "
        f"Mean F1: {mean_f1:.4f}, "
        f"Std F1: {std_f1:.4f}, "
        f"Global Best F1: {global_best_f1:.4f} "
        f"at threshold {global_best_thr:.4f}"
    )

    # ---- Update best seed ----
    if mean_f1 > best_f1_mean:
        best_seed = seed
        best_f1_mean = mean_f1
        best_f1_std = std_f1
        best_global_f1_final = global_best_f1
        best_global_thr_final = global_best_thr

# ===================== RESULT =====================
print(f"\n===== Best seed = {best_seed} =====")
print(f"Mean F1 = {best_f1_mean:.4f}, Std F1 = {best_f1_std:.4f}")
print(f"Global Best F1 = {best_global_f1_final:.4f} at threshold {best_global_thr_final:.4f}")
```
